```python
from random import randint
import numpy as np
from numba import njiti
```

```python
# @njit
def run(N):
    N2 = N*N
    d = np.zeros((N2,N2), dtype=np.int32)
    V = np.zeros((N2,N2), dtype=np.int32)
    U = np.zeros((N2,N2), dtype=np.int32)
    dU = np.zeros((N2,N2), dtype=np.int32)
    iteration = 0

    for i in range(N2):
        for j in range(i+1,N2):
            a = (i//N, i%N)
            b = (j//N, j%N)
            if ((a[0]-b[0])**2 + (a[1]-b[1])**2) == 5:
                d[i][j] = 1
                V[i][j] = randint(0,1)
    #             print("{} -> {}".format(a,b))


    while True:
        sum_du = 0
        iteration += 1

        for i in range(N2):
            for j in range(i+1,N2):
                dU[i][j] = 4
                for k in range(N2):
                    l1 = min(i,k)
                    l2 = min(j,k)
                    u1 = max(i,k)
                    u2 = max(j,k)
                    dU[i][j] -= ((d[l1][u1] * V[l1][u1]) + (d[l2][u2] * V[l2][u2]))

        for i in range(N2):
            for j in range(i+1,N2):
                sum_du += abs(dU[i][j] * d[i][j])
                U[i][j] += dU[i][j] * d[i][j]
                if d[i][j] != 0:
                    print(U[i][j], end = ' ')
        print()

        if sum_du == 0:
#             print(iteration)
            break;
            
        if iteration % 10000 == 0:
            print(sum_du)
            print(iteration//10000)

        for i in range(N2):
            for j in range(i+1,N2):
                U_val = U[i][j]
                if U_val > 3:
                    V[i][j] = 1
                elif U_val < 0:
                    V[i][j] = 0
        
        

    soln = []

    for i in range(N2):
        for j in range(i+1,N2):
            if V[i][j] == 1:
                soln.append((i,j))

    print(len(soln))
    return soln
```

```python
N = 6
soln = run(N)
print(soln)
solution = np.zeros((N*N), dtype=np.int32)
solution[0] = 1
cur = 0
for i in range(1,(N*N)):
    idx = i + 1
    for l in soln:
        if cur in l:
            other = l[0] if l[1] == cur else l[1]
            if solution[other] == 0:
                solution[other] = idx
                cur = other
                break
            
for i in range(N):
    for j in range(N):
        print(str(solution[i*N+j]).rjust(2), end= ' ')
    print()
print()
```
