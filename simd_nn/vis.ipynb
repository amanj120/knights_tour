```python
from ctypes import *
from os import path
from PIL import Image
import numpy as np
so_path = path.abspath("knights.so")
func = CDLL(so_path)
run = func.run
d_in = (c_int*66)
```

```python
gif_images = []

r_in = d_in()
run(r_in)
data = np.array(list(r_in), dtype=np.int32)
im = np.zeros((128,128,3), dtype=np.uint8);
for gi in range(10): 
    print(gi)
    for i in range(8):
        for j in range(8):
            for x in range(16):
                for y in range(16):
                    val = data[i*8+j] - 1
    #                 r = cols[val][0]*31 + 7
    #                 g = cols[val][1]*31 + 7
    #                 b = cols[val][2]*31 + 7
    #                 r = (val%4) * 85
    #                 g = ((val//4)%4)*85
    #                 b = ((val//16)%4)*85
    #                 r = 255 - (val*4)
    #                 b = (val*4)
                    im[i*16+x][j*16+y][0] = val*4
                    im[i*16+x][j*16+y][1] = val*4
                    im[i*16+x][j*16+y][2] = val*4
    im_t = Image.fromarray(im)
    print(type(im_t))
    gif_images.append(im_t)
gif_images[0].save("out.gif", save_all=True,  append_images=gif_images[1:], optimize=False, duration=40, loop=0)
```
