```python
import re
```

```python
re_str = "([ ]{0,1}[0-9]{1,2}[ ]){8}"
```

```python
f = open("all_results.txt", "r")
r = f.read()
f.close()
```

```python
lines = r.split("\n")
paths = []

for l in lines:
    if re.search(re_str, l) is not None:
        paths.append(l)
```

```python
paths
```

```python
fulls = []
for i in range(0, len(paths), 8):
    s = ""
    for j in range(8):
        s += paths[i+j]
    s2 = s.split(' ')
    s3 = []
    for s4 in s2:
        try:
            s3.append(int(s4))
        except:
            pass
    assert(len(set(s3)) == 64)
    fulls.append(s3)
    
```

```python
fulls[0]
```

```python
strs = []
for fps in fulls:
    if fps[0] != 0:
        add = 64 - fps[0]
        for i in range(64):
            fps[i] = (fps[i]+add)%64
    strs.append(" ".join([str(asdf) for asdf in fps]))
```

```python
strs[0]
```

```python
for string in strs:
#     print(string)
    assert(len(string) == 181)
    assert string[0] == '0'
    tmpl = [int(p) for p in string.split(' ')]
    assert len(tmpl) == 64
    assert len(set(tmpl)) == 64
    assert min(tmpl) == 0
    assert max(tmpl) == 63
#     print(tmpl)
    for i in range(64):
        assert tmpl.count(i) == 1
```

```python
setstrs = set(strs)
from random import randint
finalstrs = list(setstrs)
```

```python
f2 = open("compiled.txt", "w")
while(len(finalstrs) != 0):
    f2.write("{}\n".format(finalstrs.pop(randint(0,len(finalstrs)-1))))
f2.close()
```
